```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("video_games_sales.csv")
```

```python
# Display the first few rows of the dataset
print("First 5 rows of the dataset:")
print(df.head())

# Get information about the dataset
print("\nDataset Information:")
print(df.info())

# Generate summary statistics
print("\nSummary Statistics:")
print(df.describe())
```

```python
# Identify key variables to visualize (example)
key_variables = ['Global_Sales', 'NA_Sales', 'EU_Sales']
print("\nKey Variables to Visualize:")
print(key_variables)
```

```python
# Categorical Variable Visualization: Platform
plt.figure(figsize=(10, 6))
sns.countplot(x='Platform', data=df)
plt.title('Frequency of Video Game Platforms')
plt.xticks(rotation=45)
plt.show()

# Optional: Pie chart for Platforms
platform_counts = df['Platform'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(platform_counts, labels=platform_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Video Game Platforms')
plt.show()

# Numerical Variable Visualization: Global Sales
plt.figure(figsize=(10, 6))
sns.histplot(df['Global_Sales'], bins=30, kde=True)
plt.title('Distribution of Global Sales')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Global_Sales'])
plt.title('Boxplot of Global Sales')
plt.show()

plt.figure(figsize=(10, 6))
sns.violinplot(x=df['Global_Sales'])
plt.title('Violin Plot of Global Sales')
plt.show()
```

```python
# Categorical vs Numerical: Platform vs Global Sales
plt.figure(figsize=(12, 6))
sns.boxplot(x='Platform', y='Global_Sales', data=df)
plt.title('Global Sales by Platform')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 6))
sns.violinplot(x='Platform', y='Global_Sales', data=df)
plt.title('Global Sales Distribution by Platform')
plt.xticks(rotation=45)
plt.show()

# Numerical vs Numerical: NA Sales vs EU Sales
plt.figure(figsize=(10, 6))
sns.scatterplot(x='NA_Sales', y='EU_Sales', data=df)
plt.title('NA Sales vs EU Sales')
plt.show()

plt.figure(figsize=(10, 6))
sns.regplot(x='NA_Sales', y='EU_Sales', data=df)
plt.title('Regression Line: NA Sales vs EU Sales')
plt.show()

# Calculate the correlation coefficient
correlation = df['NA_Sales'].corr(df['EU_Sales'])
print(f"Correlation Coefficient between NA Sales and EU Sales: {correlation:.2f}")
```

```python
# Multivariate Visualization with Faceting: Global Sales by Platform and Genre
g = sns.catplot(x='Platform', y='Global_Sales', hue='Genre', data=df, kind='box', height=6, aspect=2)
g.set_xticklabels(rotation=45)
g.fig.suptitle('Global Sales by Platform and Genre')
plt.show()

# FacetGrid example: Scatterplot of NA Sales vs EU Sales by Genre
facet = sns.FacetGrid(df, col='Genre', col_wrap=4, height=4)
facet.map(sns.scatterplot, 'NA_Sales', 'EU_Sales')
facet.fig.suptitle('NA Sales vs EU Sales by Genre')
plt.show()
```

```python
# Univariate Plot 1: Distribution of Global Sales
plt.figure(figsize=(10, 6))
sns.histplot(df['Global_Sales'], bins=30, kde=True)
plt.title('Distribution of Global Sales')
plt.xlabel('Global Sales (in millions)')
plt.ylabel('Frequency')
plt.show()

# Commentary: This plot shows the distribution of global sales across all video games. Most games have lower sales, with a few having exceptionally high sales.

# Univariate Plot 2: Frequency of Platforms
plt.figure(figsize=(12, 6))
sns.countplot(x='Platform', data=df)
plt.title('Frequency of Video Game Platforms')
plt.xlabel('Platform')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()


# Bivariate Plot 1: NA Sales vs EU Sales
plt.figure(figsize=(10, 6))
sns.scatterplot(x='NA_Sales', y='EU_Sales', data=df)
plt.title('NA Sales vs EU Sales')
plt.xlabel('NA Sales (in millions)')
plt.ylabel('EU Sales (in millions)')
plt.show()


# Bivariate Plot 2: Boxplot of Global Sales by Genre
plt.figure(figsize=(12, 6))
sns.boxplot(x='Genre', y='Global_Sales', data=df)
plt.title('Global Sales by Genre')
plt.xlabel('Genre')
plt.ylabel('Global Sales (in millions)')
plt.xticks(rotation=45)
plt.show()


# Multivariate Visualization 1: Global Sales by Platform and Genre
g = sns.catplot(x='Platform', y='Global_Sales', hue='Genre', data=df, kind='box', height=6, aspect=2)
g.set_xticklabels(rotation=45)
g.fig.suptitle('Global Sales by Platform and Genre')
plt.show()


# Multivariate Visualization 2: FacetGrid of NA Sales vs EU Sales by Genre
facet = sns.FacetGrid(df, col='Genre', col_wrap=4, height=4)
facet.map(sns.scatterplot, 'NA_Sales', 'EU_Sales')
facet.fig.suptitle('NA Sales vs EU Sales by Genre')
plt.show()

```
